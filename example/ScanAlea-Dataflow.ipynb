Run the ViewMapOnCan node
--------------------------

The id of the **ViewMapOnCan** is 21

1. Load the PackageManager
2. Get the dataflow
3. Instantiate it
4. Run the ViewMapOnCan node
5. Display the result

```python
# First load the Package Manager
pm  = pkgmanager.PackageManager(verbose=False)
dataflow_factory = pm[pkname][node]

#instantiate
df = dataflow_factory.instantiate()

# run the dataflow
df.eval_as_expression(21)

# Get the result
scene = df.node(21).output(1)
scene = Scene([scene])

```
